SCALLING

```python
df.columns
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
```

```python
num_cols = ['bedrooms', 'bathrooms', 'sqft_living', 'sqft_lot', 'floors',
       'waterfront', 'view', 'condition', 'sqft_above', 'sqft_basement',
       'yr_built', 'yr_renovated', 'city', 'statezip', 'year', 'month', 'day']
```

```python
df[num_cols] = scaler.fit_transform(df[num_cols])
```

```python
df.head()
```

```python
x = df.drop(columns=['price'],axis=1)
y = df['price']
```

```python
from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test = train_test_split(x, y, test_size = 0.2,random_state = 42)
```

MODEL TRAINING

```python
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression()
model.fit(x_train,y_train)
```

```python
y_pred = model.predict(x_test)
y_pred
```

```python
y_test
```

```python
from sklearn.metrics import r2_score
```

```python
r2 = r2_score(y_test,y_pred)
r2
```

```python
n = x_test.shape[0]
p = x_test.shape[1]
adj_r2 = 1-((1-r2) * (n-1)) / (n-p-1)
adj_r2
```

```python
# so i can see clearly my training data and testing data both give a very bad r2 so, my data was under feet i am going to uze xgboost for get the better r2
```

```python
from xgboost import XGBRegressor
```

```python
xgb_model = XGBRegressor(n_estimators = 100,learning_rate = 0.1,max_depth = 8,
                         eval_metric = 'mlogloss',random_state = 42)
```

```python
xgb_model.fit(x_train,y_train)
```

```python
from sklearn.metrics import r2_score

pred = xgb_model.predict(x_test)

score = r2_score(y_test, pred)

print(score)
```
